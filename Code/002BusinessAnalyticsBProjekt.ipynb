```python
df['turnover_day'] = df['sales']*df['price']
```

```python
df_plot = df[['week_id', 'sales', 'price', 'turnover_day']].groupby(by='week_id').agg({'sales': 'sum', 'price': 'mean', 'turnover_day': 'sum'})
#df_plot.reset_index(inplace=True)
df_plot
```

```python
sns.barplot(x=df_plot.index, y='turnover_day', data=df_plot)
```

```python
sns.stripplot(
    data=df_plot,
    y='turnover_day'
)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=False)
#fig.suptitle('')

sns.boxplot(data=df_plot, x='sales', showmeans=True, showfliers=False, ax=axes[0])
axes[0].set_title('sales per week')

sns.boxplot(data=df_plot, x='price', showmeans=True, showfliers=False, ax=axes[1])
axes[1].set_title('Avg Price per week')

sns.boxplot(data=df_plot, x='turnover_day', showmeans=True, showfliers=False, ax=axes[2])
axes[2].set_title('turnover per week')
```

```python
df_plot2 = df[['date', 'turnover_day']].groupby(by='date').sum()
df_plot2.reset_index(inplace=True)
df_plot2['year'] = df_plot2['date'].dt.year
df_plot2['month'] = df_plot2['date'].dt.month

plot = sns.lineplot(
    data=df_plot2,
    x='month',
    y='turnover_day',
    hue='year',
    estimator=sum,
    ci=None,
    palette='deep'
).set(title='Turnover per Month')
```
